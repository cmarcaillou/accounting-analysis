```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = '/content/Invoices (27).xlsx'
df = pd.read_excel(file_path)

# Create a copy of the dataframe for modification
df_demo = df.copy()
```

```python
df_demo.shape
```

```python
for col in df_demo:
  print(col)
  print(len(df_demo[col].unique()))
```

```python

# Add a small random perturbation to all numeric columns
for column in ['Ordered amount (tax incl.)', 'Delivered amount (tax excl.)',
       'Invoiced amount (tax excl.)']:
    df_demo[column] = df_demo[column] + np.random.uniform(low=-1000, high=1000, size=len(df_demo))

# Anonymize the 'Supplier' and 'Bill to' columns
def anonymize_column(column, name):
    unique_values = df_demo[column].unique()
    anonym_map = {original: f"{name}{i+1}" for i, original in enumerate(unique_values)}
    df_demo[column] = df_demo[column].map(anonym_map)

# Apply anonymization
anonymize_column('Supplier', 'Supplier ')
anonymize_column('Bill to', 'Store ')
anonymize_column('Legal Company', 'Company ')

# Save the modified dataframe to a new Excel file
output_path = '/content/Demo_Invoices.xlsx'
df_demo.to_excel(output_path, index=False)
```
